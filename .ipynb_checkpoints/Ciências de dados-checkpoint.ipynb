```python
import pandas as pd
import numpy as np

rank_youtube_df = pd.read_csv(r'C:\Users\deivi\Desktop\Mentoria\Segundo Projeto Mentoria\topSubscribed.csv', sep=',', encoding='ISO-8859-1')
# primeira coisa que vou fazer com esses dados é tratar pra deixar minhas informações mais confiáveis
display(rank_youtube_df)

#tratar as colunas, Subscribers, Video Views, Video Count(estão aparecendo como objeto. preciso trasnformar em números)
rank_youtube_df['Subscribers'] = rank_youtube_df['Subscribers'].str.replace(',','')
rank_youtube_df['Subscribers'] = rank_youtube_df['Subscribers'].astype(int)
rank_youtube_df['Video Views'] = rank_youtube_df['Video Views'].str.replace(',','')
rank_youtube_df['Video Views'] = pd.to_numeric(rank_youtube_df['Video Views'], errors = 'coerce')
rank_youtube_df['Video Count'] = rank_youtube_df['Video Count'].str.replace(',','')
rank_youtube_df['Video Count'] = pd.to_numeric(rank_youtube_df['Video Count'], errors = 'coerce')

#info utilizo para verificar se tenho algum item vazio
print(rank_youtube_df.info())

#df.isna().any() ou sum() tbm trata item vazio
print(rank_youtube_df.isna().sum())
```

```python
#qual categoria mais assistida
categoria_assistida = rank_youtube_df['Category'].value_counts()
display(categoria_assistida)
categoria_assistida[:5].plot(yticks=range(0, 300, 20))
```

```python
#qual canal tem mais inscritos
rank_youtube_df = rank_youtube_df.sort_values('Subscribers', ascending=False)

display(rank_youtube_df[:5][['Youtube Channel', 'Subscribers', 'Category', 'Started']])

rank_youtube_df[:5][['Youtube Channel', 'Subscribers', 'Category', 'Started']].plot.bar(x='Youtube Channel',y='Subscribers')
```

```python
#qual canal tem mais número de visualizações
rank_youtube_df = rank_youtube_df.sort_values('Video Views', ascending=False)

display(rank_youtube_df[:5][['Youtube Channel', 'Video Views', 'Category', 'Started']])

rank_youtube_df[:5][['Youtube Channel', 'Video Views', 'Category', 'Started']].plot.bar(x='Youtube Channel',y='Video Views')
```

```python
#qual canal creceu mais rápido
from datetime import date
ano_da_df = rank_youtube_df[:5]['Started']
ano_atual = date.today().year
ano_dif = ano_atual - ano_da_df

media_cresc_insc = (rank_youtube_df['Subscribers'] / ano_dif) / 1000000

media_df = pd.DataFrame(media_cresc_insc)

rank_youtube_df['Média de Crescimento'] = media_df

rank_youtube_df = rank_youtube_df.sort_values('Média de Crescimento', ascending=False)

display(rank_youtube_df[:5][['Youtube Channel', 'Subscribers', 'Category', 'Média de Crescimento']])

rank_youtube_df[:5][['Youtube Channel', 'Subscribers', 'Category', 'Média de Crescimento']].plot.bar(x='Youtube Channel',y='Média de Crescimento')
```
